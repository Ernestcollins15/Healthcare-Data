```python
#Necessary Libraries
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
import os
```

```python
#First check the current working directory
pwd=os.getcwd()
pwd
```

```python
#Creating a filepath to your directory where your the dataset is
filepath_diag=os.path.join(pwd,"Diagnosis.csv")
filepath_inv = os.path.join(pwd,"Invoice.csv")
filepath_visit = os.path.join(pwd,"Visit.csv")
```

```python
penda_diagnosis = pd.read_csv(filepath_diag)
penda_visit = pd.read_csv(filepath_visit)
penda_invoice = pd.read_csv(filepath_inv)
```

```python
penda_visit
```

```python
penda_invoice
```

```python
penda_diagnosis
```

```python
penda_visit.merge(penda_invoice, left_on= 'VisitId', right_on= 'VisitCode')
```

```python
visit_diag = penda_visit.merge(penda_diagnosis, left_on= 'VisitId', right_on= 'VisitCode')
visit_diag
```

```python
visit_diag_inv = visit_diag.merge(penda_invoice, left_on= 'VisitCode', right_on= 'VisitCode')
visit_diag_inv
```

```python
visit_diag_inv.groupby(["Payor","VisitCategory"]).sum()
```

```python
#Those who both visited in person and online
count_both = visit_diag_inv.query("VisitCategory == 'Telemedic Visit' & VisitCategory == 'In-person Visit'")['PatientCode'].count()
count_both 
```

```python
visit_diag_inv.groupby(["Diagnosis","MedicalCenter"]).sum()
```

```python
#most profitable payor
grouped_data = visit_diag_inv.groupby(["Diagnosis", "MedicalCenter"]).sum()
sorted_data = grouped_data.sort_values(by="Diagnosis", ascending=False)
sorted_data 
```

```python
groupeddata = visit_diag_inv.groupby(["Payor", "Amount_adjusted"]).sum()
sorted_ = groupeddata.sort_values(by="Payor", ascending=False)
print(sorted_)
```
